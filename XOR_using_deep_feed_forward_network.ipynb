```python
import numpy as np
```

```python
def unit_step(v):
    if v >= 0:
        return 1
    else:
        return 0
```

```python
def perceptron(x, w, b):
    v = np.dot(x, w) + b
    y = unit_step(v)
    return y
```

```python
def not_gate(x):
    wnot = -1
    bnot = 0.5
    return perceptron(x, wnot, bnot)
```

```python
def and_gate(x):
    wand = np.array([1, 1])
    band = -1.5
    return perceptron(x, wand, band)
```

```python
def or_gate(x):
    wor = np.array([1, 1])
    bor = -0.5
    return perceptron(x, wor, bor)
```

```python
def xor_gate(x):
    s1 = and_gate(x)
    s2 = not_gate(s1)
    s3 = and_gate(np.array([x[0], s2]))
    s4 = and_gate(np.array([s2, x[1]]))
    y = or_gate(np.array([s3, s4]))
    return y
```

```python
test1 = np.array([0, 0])
test2 = np.array([0, 1])
test3 = np.array([1, 0])
test4 = np.array([1, 1])
print("XOR({}, {}) = {}".format(test1[0], test1[1], xor_gate(test1)))
print("XOR({}, {}) = {}".format(test2[0], test2[1], xor_gate(test2)))
print("XOR({}, {}) = {}".format(test3[0], test3[1], xor_gate(test3)))
print("XOR({}, {}) = {}".format(test4[0], test4[1], xor_gate(test4)))
```
